```python
import pandas as pd

# Load the Excel file
df_threat = pd.read_excel("taxon_threat_impact.xlsx")

# Preview first rows
df_threat.head()
```

```python
# Check info (columns, dtypes, nulls)
df_threat.info()

# Check missing values
df_threat.isna().sum()

# Get a quick overview of categorical columns
for col in df_threat.select_dtypes(include="object").columns:
    print(f"\nUnique values in {col}:")
    print(df_threat[col].unique()[:10])  # show first 10 unique values
```

```python
# Apply filter rules
df_filtered = df_threat[
    (df_threat["IUCN threat level 1 description"] == "Invasive and other problematic species, genes and diseases") &
    (df_threat["IUCN threat level 2 description (or finest level available)"] == "Invasive non-native/alien species/diseases")
].reset_index(drop=True)

df_filtered.info()
df_filtered.head()
```

```python
df_filtered["IUCN threat level 3 description (or finest level available)"].unique()
```

```python
# Define keywords to search for 10 specific most common invasive species
keywords = [
    "fox", "rabbit", "pig", "common myna", "cane toad", 
    "lantana", "bitou bush", "gorse", "buffel grass", "gamba grass"
]

# regex pattern (case-insensitive)
pattern = "|".join(keywords)

# Filter rows
df_filtered_species = df_filtered[
    df_filtered["IUCN threat level 3 description (or finest level available)"]
    .str.contains(pattern, case=False, na=False)
].reset_index(drop=True)

# Recheck for unique species names
df_filtered_species["IUCN threat level 3 description (or finest level available)"].unique()
```

```python
df_filtered_species.info()
```

```python
df_filtered_species["Group"].unique()
```

```python
# Mapping dictionary: keyword -> clean species name
species_map = {
    "fox": "Red Fox",
    "rabbit": "European Rabbit",
    "pig": "Feral Pig",
    "common myna": "Common Myna",
    "cane toad": "Cane Toad",
    "lantana": "Lantana",
    "bitou bush": "Bitou Bush",
    "gorse": "Gorse",
    "buffel grass": "Buffel Grass",
    "gamba grass": "Gamba Grass"
}

# Function to assign species name
def map_species(desc):
    for keyword, clean_name in species_map.items():
        if pd.notna(desc) and keyword.lower() in desc.lower():
            return clean_name
    return None

# Apply function to create new column
df_filtered_species["inv_sp_name"] = df_filtered_species[
    "IUCN threat level 3 description (or finest level available)"
].apply(map_species)

# Quick check
df_filtered_species[["IUCN threat level 3 description (or finest level available)", "inv_sp_name"]].head(20)
```

```python
df_filtered_species.head()
```

```python
# Define the columns you want to keep
cols_to_keep = [
    "Species name adjusted", "Species name", "Group", 
    "Broad level threat", "Scope", "Severity", 
    "Timing score", "Sum of timing, scope, severity", 
    "Impact score", "inv_sp_name"
]

# Filter the DataFrame
df_final = df_filtered_species[cols_to_keep].reset_index(drop=True)

# Quick checktax
df_final.head()
```

```python
df_final.to_csv("taxon_threat_impact_cleaned.csv", index=False)
```
